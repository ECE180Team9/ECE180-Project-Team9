```python
from pylab import *
from matplotlib.widgets import Slider, Button, RadioButtons
import matplotlib.image as mpimg

fig,ax = subplots()
subplots_adjust(left=0.25, bottom=0.25)
ax.imshow(mpimg.imread('img17.png'))


rax = axes([0.025, 0.5, 0.15, 0.15])
radio = RadioButtons(rax, ('2017', '2016', '2015'), active=0)
def colorfunc(label):
    if label == '2017' :
        ax.imshow(mpimg.imread('img17.png'))
    if label == '2016' :
        ax.imshow(mpimg.imread('img16.png'))
    if label == '2015' :
        ax.imshow(mpimg.imread('img15.png'))
radio.on_clicked(colorfunc)

show()
```
